```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
#lectura de documento 
titanic = pd.read_csv("../../datasets/titanic/train.csv")
titanic.head()
```

```python
#revision de data info/cantidad de datos
print(titanic.info())
print(f"cantidad de datos : {titanic.shape}")
```

```python
#revision de datos nulos 
print(titanic.isnull().sum())
```

## Eliminar valores nulos

```python
titanic_clean = titanic.dropna(subset=['Age','Cabin','Embarked'])
```

## EDA

```python
plt.title("Supervivientes al titanic por Sexo")
sns.countplot(data=titanic_clean,x='Survived',hue='Sex')
plt.show()

por_sexo_supervivientes = titanic_clean.loc[(titanic_clean['Survived']==1),]
por_sexo_nosupervivientes = titanic_clean.loc[(titanic_clean['Survived']==0),]

print("supervivientes")
print(por_sexo_supervivientes.Sex.value_counts() / por_sexo_supervivientes.shape[0])
print("no supervivientes")
print(por_sexo_nosupervivientes.Sex.value_counts() / por_sexo_nosupervivientes.shape[0])
```

```python
#rangos de edad entre supervivientes hombre
fig,ax = plt.subplots(1,2,figsize=(10,5))
sns.histplot(data=por_sexo_supervivientes,x='Age',hue='Sex',ax=ax[0])
sns.histplot(data=por_sexo_nosupervivientes,x='Age',hue='Sex',ax=ax[1])

#todo: Agregar titulos y obtener intervalos de confianza de un 95%.
```

<b>comentarios</b>: a partir del grafico observamos lo siguiente: 
<li>Para los supervivientes el 66.6% fueron mujeres entre y 33.3% hombres</li>

## Transformacion de datos y entranamiento del modelo

```python
#seleccionar variables importantes
features = ['Pclass','Sex','Age','Fare','Embarked','SibSp']
target = 'Survived'

#X e y 
X = titanic_clean[features]
y = titanic_clean[target]

#aplicar labelEncoder 
from sklearn.preprocessing import LabelEncoder,StandardScaler

encoder = LabelEncoder()
scaler = StandardScaler()

#Escalar valores
## seleccionar variables categoricas 
columnas_categoricas = X.select_dtypes("object").columns
for column in columnas_categoricas:
    X[column] = encoder.fit_transform(X[column])

# Entrenamiento modelo
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix

lg = LogisticRegression().fit(X,y)
train_predicts = lg.predict(X)
train_metrics = classification_report(y,train_predicts)

print(train_metrics)

#matrix de confusion
cnf_matrix = confusion_matrix(train_predicts,y)

plt.title("train confussion matrix")
sns.heatmap(cnf_matrix,annot=True,cmap='Blues',fmt='g')
plt.show()
```

## Interpretando resultados

Entendiendo la matri de confusion.

<br>
comentarios : En este primer caso, la matriz de confusion nos indica lo siguiente:
<li>41 casos el modelo clasifico falso cuando eran falso</li>
<li>en 19 casos era falso pero se clasifico como verdadero</li>
<li>en 21 casos era verdadero pero se calsifico como falso</li>
<li>en 102 casos era verdadero y se clasifico como verdadero</li>
